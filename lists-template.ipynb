## Working with lists

```python
# declare an empty string
bob = ''
# declare a list
beatles = []
```

### What kind of variable is this?

### Populate and print a list

```python
beatles = ["John", "Paul", "George", "Ringo", "Stuart"]
#print the whole list
print(beatles)
```

### Calculate and print the length of a list

```python
# calculate number of beatles
number_of_beetles = len(beatles)
# print number of beatles
print(f"There are {number_of_beetles} members of the beatles.")
```

### Print an item in a list 
   
   _see PowerPoint slides_

```python
#remind ourselves the list
print(beatles)

# print the first beatle
print(beatles[0])

# print the last Beatle
print(beatles[4])

#other ways to print last beatle
print(beatles[-1])
print(beatles[len(beatles)-1])

#many ways to print george
print(beatles[2]) #george indexing from start
print(beatles[-3]) #george indexing from end
```

### Remove an item from a list

```python
# Remove the last item in the list (bye, Stuart)
stu = beatles.pop()
# What happened to the Beatles?
print(stu)
print(beatles)
```

### Make a new list

```python
# make a list called 'stones' with the Rolling Stones (Mick, Ron, and Keith)
stones = ["Mick", "Keith", "Ron"]
```

### Add two lists

```python
# add two lists together
print("the two groups seperately: {}, {}".format(beatles, stones))
print()

#add the lists together
supergroup = beatles + stones

# print our new list
print(f"{len(supergroup)} members in our new band: {supergroup}")
```

### Growing a list with `extend()`

```python
# print our lists
print("jackson5:", jackson5)
print("destiny:", destiny)
# use extend() to add Destiny's Child to the Jackson 5
```

```python

# extend() modifies a list in place, it doesn't return a value (i.e., a new list)
jackson5.extend(destiny)

#here is what is is doing
#jackson5 = jackson5 + destiny

print(f"Here is the extended Jackson5: {jackson5}")
```

### Growing a list with `append()`
